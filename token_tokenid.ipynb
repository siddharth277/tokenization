```python
import tiktoken
```

```python
def generatetokenid(text : str, model : str) :
    tokenizer = tiktoken.encoding_for_model(model_name = model)
    return tokenizer.encode(text)

    
```

```python
text = input()
model = input()
print(text, model)
resultids = generatetokenid(text, model)
print(resultids)
```

```python
tokenizer = tiktoken.encoding_for_model(model_name = model)
```
